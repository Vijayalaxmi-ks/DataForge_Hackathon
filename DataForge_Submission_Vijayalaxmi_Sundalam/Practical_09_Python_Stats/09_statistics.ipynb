```python
import pandas as pd
import numpy as np

# Load our processed dataset from Practical 8
df = pd.read_csv("../../01_data/processed/dataforge_cleaned.csv")

# Generate and export numerical summaries (Task 2)
numeric_summary = df.describe().T
numeric_summary.to_csv("../../04_python/outputs/tables/numeric_summary.csv")

# Create missing-value percentage table and export (Task 3)
missing = (df.isna().mean() * 100).sort_values(ascending=False).to_frame(name="Missing_Percent")
missing.to_csv("../../04_python/outputs/tables/missing_percent.csv")

print("Baseline descriptive tables exported successfully!")
```

```python
# Generate at least 4 distinct grouped tables using .groupby()
grp_category = df.groupby('Category', as_index=False)[['Sales', 'Total Profit']].sum()
grp_region = df.groupby('Region', as_index=False)[['Sales', 'Total Profit']].mean()
grp_segment = df.groupby('Segment', as_index=False)['Shipping Delay'].mean()
grp_shipmode = df.groupby('Ship Mode', as_index=False)['Sales'].count().rename(columns={'Sales': 'Order_Volume'})

# Export grouped insights
grp_category.to_csv("../../04_python/outputs/tables/grp_category.csv", index=False)
grp_region.to_csv("../../04_python/outputs/tables/grp_region.csv", index=False)
grp_segment.to_csv("../../04_python/outputs/tables/grp_segment.csv", index=False)
grp_shipmode.to_csv("../../04_python/outputs/tables/grp_shipmode.csv", index=False)

print("All 4 grouped calculation tables exported successfully!")
```

```python
# Aggregate by Sub-Category to get net volume metrics
subcat_summary = df.groupby('Sub-Category', as_index=False)['Sales'].sum()

# Extract top 10 and bottom 10 ranking frames
top_10_subcats = subcat_summary.sort_values(by='Sales', ascending=False).head(10)
bottom_10_subcats = subcat_summary.sort_values(by='Sales', ascending=True).head(10)

# Export rankings
top_10_subcats.to_csv("../../04_python/outputs/tables/top_10_subcats.csv", index=False)
bottom_10_subcats.to_csv("../../04_python/outputs/tables/bottom_10_subcats.csv", index=False)

print("Ranking profiles successfully compiled and written to tables folder.")
```

```python
# 1. Filter to include only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# 2. Generate a correlation matrix using .corr()
correlation_matrix = numerical_df.corr()
correlation_matrix.to_csv("../../04_python/outputs/tables/correlation_matrix.csv")

# 3. Check the skewness of the 'Sales' column
sales_skew = df['Sales'].skew()

print("Correlation Matrix saved!")
print(f"The skewness coefficient of the Sales column is: {sales_skew:.4f}")
```
