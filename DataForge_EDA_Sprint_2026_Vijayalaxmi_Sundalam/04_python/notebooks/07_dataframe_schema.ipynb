```python
import pandas as pd

# Use ../../ to step out of 'notebooks' and '04_python' folders back to the main workspace root
df = pd.read_csv("../../01_data/raw/dataset_raw.csv")

# Print basic shape and memory allocation traces
print(f"Dataset Shape: {df.shape}")
print("-" * 40)
df.info()
```

```python
# View top records
df.head()

# View statistical dispersion summaries
df.describe(include='all')

# Check raw structural missing counts
df.isnull().sum()
```

```python
# Build structural inspection audit frame as defined in image_1aedcd.png
schema = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype) for c in df.columns],
    "missing": [df[c].isna().sum() for c in df.columns],
    "unique": [df[c].nunique() for c in df.columns],
    "sample": [df[c].dropna().iloc[0] if df[c].dropna().shape[0] else "" for c in df.columns]
})

# Use ../../ to step out of notebooks/ and 04_python/ back to the root workspace directory
schema.to_csv("../../04_python/outputs/tables/schema_audit.csv", index=False)
print("Schema audit exported successfully!")
```
